# Image Binarization

When binarizing an image, we produce an image that has only two values: True and False. Often, images are also called if they contain the value `0`, e.g. for background, and any other value for foreground.

```python
import numpy as np
from skimage import data
from skimage.io import imread
from skimage.filters import threshold_otsu
import matplotlib.pyplot as plt
```

We use this example image of nuclei.

```python
image_nuclei = imread('data/mitosis_mod.tif')

plt.imshow(image_nuclei, cmap='gray')
plt.colorbar()
```

## Image Thesholding
The most common binarization technique is thresholding. We _apply a threshold_ to determine which pixel lie above a certain pixel intensity and which are below.

```python
image_binary = image_nuclei > 125
```

```python
plt.imshow(image_binary, cmap='gray')
plt.colorbar()
```

As choosing a manual threshold introduces human bias, it might be better to apply an automated thresholding algorithm such as [Otsu's method](https://s3-us-west-2.amazonaws.com/ieeeshutpages/xplore/xplore-shut-page.html).

```python
threshold = threshold_otsu(image_nuclei)
threshold
```

```python
image_otsu_binary = image_nuclei > threshold

plt.imshow(image_otsu_binary, cmap='gray')
plt.colorbar()
```
